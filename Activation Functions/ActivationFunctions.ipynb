Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

Step Function

```python
# Definition

def stepFunction(x):
    return np.where(x >= 0, 1, 0)
```

```python
# X values
x = np.linspace(-10, 10, 400)

# Step Function
y = stepFunction(x)

# Graphic
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Step Function", color='blue')
plt.title("Step Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

Sigmoid Function

```python
# Definition

def sigmoidFunction(x):
    return 1 / (1 + np.exp(-x))
```

```python
# X values
x = np.linspace(-10, 10, 400)

# Sigmoid Function
y = sigmoidFunction(x)

# Graphic
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Sigmoid Function", color='red')
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

Tanh Function

```python
# Definition

def tanhFunction(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
# X values
x = np.linspace(-10, 10, 400)

# Sigmoid Function
y = tanhFunction(x)

# Graphic
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Tanh Function", color='green')
plt.title("Tanh Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

ReLu Function

```python
# Definition

def reluFunction(x):
    return np.maximum(0, x)
```
